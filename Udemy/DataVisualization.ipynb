```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```
%matplotlib inline
```

```
x = np.arange(0,11)
y = x**2
```

```
x
```

```
y
```

```
plt.plot(x,y,'r--')
plt.xlim(0,10)
plt.ylim(0,100)
plt.title("X VS Y")
plt.xlabel('X')
plt.ylabel('Y')
```
